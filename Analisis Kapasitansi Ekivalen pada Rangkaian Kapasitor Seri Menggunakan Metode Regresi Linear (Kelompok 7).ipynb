### 2. Metode Komputasi

Metode komputasi dilakukan menggunakan bahasa pemrograman Python dengan metode regresi linear kuadrat terkecil (*least squares regression*).

Persamaan regresi linear yang digunakan adalah:

$$
y = a + bx
$$

dengan:

$$
a = \text{intercept}
$$

$$
b = \text{slope}
$$

Nilai slope dari regresi linear digunakan untuk menentukan kapasitansi percobaan melalui persamaan:

$$
b = -\frac{1}{C}
$$

sehingga:

$$
C = -\frac{1}{b}
$$

Hasil regresi linear dari data percobaan menghasilkan:

$$

$$

maka:

$$
C = 109.216 \,\mu F
$$

Nilai koefisien determinasi yang diperoleh adalah:

$$
R^2 = 0.9940
$$

Nilai tersebut menunjukkan bahwa hasil regresi linear memiliki kecocokan yang sangat baik terhadap data percobaan.

Perbedaan hasil teori dan praktik dapat disebabkan oleh:

- serta hambatan internal komponen.

Namun, hasil regresi linear menunjukkan kecocokan yang sangat baik terhadap data percobaan dengan nilai:

$$
R^2 = 0.9940
$$

Nilai tersebut menunjukkan bahwa model regresi linear mampu merepresentasikan data percobaan dengan sangat baik.

```python
import numpy as np
import matplotlib.pyplot as plt

def least_squares_regression(x_data, y_data):
    """
    Regresi linear metode kuadrat terkecil.
    Mengembalikan intercept, slope, dan R-squared.
    """

    # Membentuk matriks regresor
    X = np.vstack([np.ones_like(x_data), x_data]).T
    y = y_data.reshape(-1, 1)

    # Menghitung parameter regresi
    beta = np.linalg.inv(X.T @ X) @ (X.T @ y)

    beta_0 = beta[0][0]   # intercept
    beta_1 = beta[1][0]   # slope

    # Prediksi y
    y_pred = beta_0 + beta_1 * x_data

    # Menghitung R²
    sse = np.sum((y_data - y_pred) ** 2)
    sst = np.sum((y_data - np.mean(y_data)) ** 2)

    r_squared = 1 - (sse / sst)

    return beta_0, beta_1, r_squared


# =====================================
# DATA PERCOBAAN
# =====================================

# Waktu (s)
t_data = np.array([0, 10, 20, 30, 40])

# Data linearisasi
y_data = np.array([
    0,
    -106189.4706,
    -207784.2820,
    -291877.1232,
    -364965.8741
])

# =====================================
# REGRESI LINEAR
# =====================================

a, b, r2 = least_squares_regression(t_data, y_data)

print("HASIL REGRESI LINEAR")
print(f"Persamaan regresi : y = {a:.4f} + ({b:.4f})x")
print(f"Intercept (a)     : {a:.4f}")
print(f"Slope (b)         : {b:.4f}")
print(f"R-squared (R²)    : {r2:.4f}")

# =====================================
# KAPASITANSI PERCOBAAN
# =====================================

# Rumus:
# b = -1/C
# C = -1/b

C_farad = -1 / b

# Konversi Farad ke mikroFarad
C_micro = C_farad * 1e6

print("\nHASIL KAPASITANSI")
print(f"Kapasitansi percobaan = {C_micro:.3f} µF")

# =====================================
# KAPASITANSI TEORITIS
# =====================================

C1 = 100   # µF
C2 = 470   # µF

# Rumus kapasitor seri
C_teori = 1 / ((1 / C1) + (1 / C2))

print("\nKAPASITANSI TEORITIS")
print(f"C teori = {C_teori:.3f} µF")

# =====================================
# PELAPORAN HASIL
# =====================================

delta_C = 4.917

print("\nPELAPORAN")
print(f"C = ({C_micro:.3f} ± {delta_C:.3f}) µF")

# =====================================
# GRAFIK REGRESI LINEAR
# =====================================

t_line = np.linspace(0, 45, 100)
y_line = a + b * t_line

plt.figure(figsize=(8,5))

plt.scatter(
    t_data,
    y_data,
    color='blue',
    label='Data Percobaan'
)

plt.plot(
    t_line,
    y_line,
    color='red',
    label=f'Garis Regresi\n$R^2$ = {r2:.4f}'
)

plt.title("Regresi Linear Kapasitansi Ekivalen")
plt.xlabel("Waktu (s)")
plt.ylabel("Nilai Linearisasi")
plt.grid(True)
plt.legend()

plt.show()

# =====================================
# GRAFIK VC TERHADAP t
# =====================================

t_vc = np.array([0, 10, 20, 30, 40])
Vc = np.array([0, 3.3, 4.4, 4.7, 4.9])
                                                                                                   p
plt.figure(figsize=(8,5))

plt.plot(
    t_vc,
    Vc,
    marker='o',
    color='green',
    label='Vc terhadap t'
)

plt.title("Grafik Vc terhadap t Rangkaian Seri")
plt.xlabel("Waktu (s)")
plt.ylabel("Tegangan Sesaat (Vc)")
plt.grid(True)
plt.legend()

plt.show()
```
