```python
sentence = "this is a rambling sentence that simply goes on and on and on and just simply will not stop that is just the way things are"

# Pecah kalimat menjadi daftar kata
words = sentence.split()

# Buat dictionary untuk menghitung kata
wordtocount = {}

for word in words:
    wordtocount[word] = wordtocount.get(word, 0) + 1

# Cetak tabel populasi kata
print("Word\tCount")
for word in wordtocount:
    print(word, "\t", wordtocount[word])

# Hitung berapa kata muncul 1x, 2x, 3x
once = 0
twice = 0
thrice = 0

for count in wordtocount.values():
    if count == 1:
        once += 1
    elif count == 2:
        twice += 1
    elif count == 3:
        thrice += 1

print("\nWords that appear once:", once)
print("Words that appear twice:", twice)
print("Words that appear thrice:", thrice)

# Verifikasi total kata
total_words = len(words)
check_total = once + 2 * twice + 3 * thrice

print("\nTotal words in sentence:", total_words)
print("Total from frequency check:", check_total)
```

```python
def read_table_loop(filename):
    xs = []
    ys = []
    with open(filename, 'r') as infile:
        for line in infile:
            line = line.strip()
            if not line:  # Skip empty lines
                continue
            parts = line.split()
            # Ensure there are at least two parts and the last two can be converted to float
            if len(parts) >= 2:
                try:
                    x_val = float(parts[-2])
                    y_val = float(parts[-1])
                    xs.append(x_val)
                    ys.append(y_val)
                except ValueError:
                    # Skip lines where the last two parts are not numbers (e.g., 'Total sum line')
                    continue
    return xs, ys

print("Defined read_table_loop function.")
```

```python
def read_table_lc(filename):
    with open(filename, 'r') as infile:
        lines = [line.strip() for line in infile if line.strip()]

    # Filter lines that have at least two parts and parse them
    parsed_numeric_parts = [
        (float(parts[-2]), float(parts[-1]))
        for line in lines
        for parts in [line.split()]
        if len(parts) >= 2 and all(part.replace('.', '', 1).isdigit() for part in parts[-2:]) # Check if parsable
    ]

    # Separate x and y values using list comprehensions
    xs = [item[0] for item in parsed_numeric_parts]
    ys = [item[1] for item in parsed_numeric_parts]

    return xs, ys

print("Defined read_table_lc function.")
```

```python
print("--- Testing read_table_loop ---")
x_loop, y_loop = read_table_loop('dummy1.dat')
print("X values (loop):", x_loop)
print("Y values (loop):", y_loop)

print("\n--- Testing read_table_lc ---")
x_lc, y_lc = read_table_lc('dummy1.dat')
print("X values (list comprehension):", x_lc)
print("Y values (list comprehension):", y_lc)
```

```python
import math

# Fungsi yang didefinisikan user
def f(x):
    return math.exp(-x**4)

# Buat daftar x dari -1 sampai 1 dengan langkah 0.1
x_values = [round(-1 + 0.1*i, 1) for i in range(21)]

# Hitung nilai f(x) dengan list comprehension
f_values = [f(x) for x in x_values]

# Cetak tabel dengan zip dan formatting
for x, fx in zip(x_values, f_values):
    print(f"{x:5.1f} {fx:10.5f}")
```
