```python
import numpy as np

# Matriks koefisien A dan vektor konstanta b
A = np.array([
    [10, -2,  1, -1],
    [-2, 11, -1,  3],
    [ 1, -1, 10, -2],
    [-1,  3, -2,  8]
], dtype=float)

b = np.array([6, 25, -11, 15], dtype=float)


# =========================================================
# 1. Metode Dekomposisi LU
# =========================================================
def lu_decomposition(A):
    n = len(A)
    L = np.zeros((n, n))
    U = np.zeros((n, n))

    for i in range(n):
        L[i, i] = 1.0

        # Menghitung elemen U
        for j in range(i, n):
            sum_u = 0
            for k in range(i):
                sum_u += L[i, k] * U[k, j]
            U[i, j] = A[i, j] - sum_u

        # Menghitung elemen L
        for j in range(i + 1, n):
            sum_l = 0
            for k in range(i):
                sum_l += L[j, k] * U[k, i]
            L[j, i] = (A[j, i] - sum_l) / U[i, i]

    return L, U


def forward_substitution(L, b):
    n = len(b)
    y = np.zeros(n)
    for i in range(n):
        sum_y = 0
        for j in range(i):
            sum_y += L[i, j] * y[j]
        y[i] = (b[i] - sum_y) / L[i, i]
    return y


def backward_substitution(U, y):
    n = len(y)
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        sum_x = 0
        for j in range(i + 1, n):
            sum_x += U[i, j] * x[j]
        x[i] = (y[i] - sum_x) / U[i, i]
    return x


def solve_lu(A, b):
    L, U = lu_decomposition(A)
    y = forward_substitution(L, b)
    x = backward_substitution(U, y)
    return L, U, x


# =========================================================
# 2. Metode Jacobi
# =========================================================
def jacobi(A, b, x0=None, tol=1e-10, max_iter=100):
    n = len(b)
    x = np.zeros(n) if x0 is None else x0.copy()
    x_new = np.zeros(n)

    for iteration in range(1, max_iter + 1):
        for i in range(n):
            sigma = 0
            for j in range(n):
                if j != i:
                    sigma += A[i, j] * x[j]
            x_new[i] = (b[i] - sigma) / A[i, i]

        error = np.linalg.norm(x_new - x, ord=np.inf)
        if error < tol:
            return x_new, iteration

        x = x_new.copy()

    return x, max_iter


# =========================================================
# 3. Metode Gauss-Seidel
# =========================================================
def gauss_seidel(A, b, x0=None, tol=1e-10, max_iter=100):
    n = len(b)
    x = np.zeros(n) if x0 is None else x0.copy()

    for iteration in range(1, max_iter + 1):
        x_old = x.copy()

        for i in range(n):
            sigma1 = 0
            sigma2 = 0

            for j in range(i):
                sigma1 += A[i, j] * x[j]

            for j in range(i + 1, n):
                sigma2 += A[i, j] * x_old[j]

            x[i] = (b[i] - sigma1 - sigma2) / A[i, i]

        error = np.linalg.norm(x - x_old, ord=np.inf)
        if error < tol:
            return x, iteration

    return x, max_iter


# =========================================================
# Menjalankan semua metode
# =========================================================
L, U, x_lu = solve_lu(A, b)
x_jacobi, iter_jacobi = jacobi(A, b)
x_gs, iter_gs = gauss_seidel(A, b)

print("=== Matriks L ===")
print(L)

print("\n=== Matriks U ===")
print(U)

print("\n=== Solusi dengan Dekomposisi LU ===")
print(f"I1 = {x_lu[0]:.6f}")
print(f"I2 = {x_lu[1]:.6f}")
print(f"I3 = {x_lu[2]:.6f}")
print(f"I4 = {x_lu[3]:.6f}")

print("\n=== Solusi dengan Metode Jacobi ===")
print(f"I1 = {x_jacobi[0]:.6f}")
print(f"I2 = {x_jacobi[1]:.6f}")
print(f"I3 = {x_jacobi[2]:.6f}")
print(f"I4 = {x_jacobi[3]:.6f}")
print(f"Jumlah iterasi = {iter_jacobi}")

print("\n=== Solusi dengan Metode Gauss-Seidel ===")
print(f"I1 = {x_gs[0]:.6f}")
print(f"I2 = {x_gs[1]:.6f}")
print(f"I3 = {x_gs[2]:.6f}")
print(f"I4 = {x_gs[3]:.6f}")
print(f"Jumlah iterasi = {iter_gs}")
```
