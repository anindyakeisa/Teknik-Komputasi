**Metode teoritis dan komputasi**

## Metode yang Digunakan

Pada penelitian ini digunakan dua metode perhitungan, yaitu metode analitik dan metode komputasi.

### 1. Metode Analitik

Proses pengisian kapasitor pada rangkaian RC dapat dimodelkan menggunakan persamaan diferensial yang diperoleh dari Hukum Kirchhoff.

Berdasarkan persamaan:

$$
V_s = V_R + V_C
$$

dengan:

$$
V_R = iR
$$

dan

$$
i = C \frac{dV_C}{dt}
$$

maka diperoleh model matematis:

$$
\frac{dV_C}{dt}
=
\frac{V_s - V_C}{RC}
$$

Persamaan diferensial tersebut menggambarkan perubahan tegangan kapasitor terhadap waktu selama proses pengisian.

Parameter yang digunakan dalam simulasi adalah:

$$
V_s = 5\ V
$$

$$
R = 100000\ \Omega
$$

$$
C = 109.21 \times 10^{-6}\ F
$$

Sehingga:

$$
RC=(100000)(109.21\times10^{-6})
$$

$$
RC=10.921
$$

---

### 2. Metode Komputasi

Metode komputasi dilakukan menggunakan bahasa pemrograman Python dengan Metode Euler untuk menyelesaikan persamaan diferensial secara numerik.

Persamaan Euler yang digunakan adalah:

$$
V_{n+1}
=
V_n
+
\Delta t
\left(
\frac{V_s-V_n}{RC}
\right)
$$

dengan:

$$
\Delta t = 1\ s
$$

dan kondisi awal:

$$
V_C(0)=0
$$

#### Iterasi Pertama

$$
V_1
=
0
+
1
\left(
\frac{5-0}{10.921}
\right)
$$

$$
V_1
=
0.4578\ V
$$

#### Iterasi Kedua

$$
V_2
=
0.4578
+
1
\left(
\frac{5-0.4578}{10.921}
\right)
$$

$$
V_2
=
0.8737\ V
$$

#### Iterasi Ketiga

$$
V_3
=
0.8737
+
1
\left(
\frac{5-0.8737}{10.921}
\right)
$$

$$
V_3
=
1.2515\ V
$$

## 7. Kesimpulan

1. Pengisian kapasitor pada rangkaian RC dapat dimodelkan menggunakan persamaan diferensial orde satu.

2. Metode Euler berhasil digunakan untuk menyelesaikan persamaan diferensial secara numerik.

3. Tegangan kapasitor meningkat secara bertahap dan mendekati tegangan sumber sebesar 5 V.

4. Hasil simulasi sesuai dengan teori pengisian kapasitor pada rangkaian RC.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter rangkaian
Vs = 5              # Tegangan sumber (Volt)
R = 100000          # Resistor (Ohm)
C = 109.21e-6       # Kapasitor (Farad)

# Parameter simulasi
dt = 1              # langkah waktu (s)
t_akhir = 40        # waktu simulasi (s)

# Array waktu
t = np.arange(0, t_akhir + dt, dt)

# Array tegangan kapasitor
Vc = np.zeros(len(t))

# Kondisi awal
Vc[0] = 0

# Metode Euler
for i in range(len(t)-1):
    dVdt = (Vs - Vc[i])/(R*C)
    Vc[i+1] = Vc[i] + dt*dVdt

# Grafik
plt.figure(figsize=(8,5))
plt.plot(t, Vc, marker='o')
plt.xlabel('Waktu (s)')
plt.ylabel('Tegangan Kapasitor (V)')
plt.title('Pengisian Kapasitor RC Menggunakan Metode Euler')
plt.grid(True)
plt.show()
```
