```python
import numpy as np

# Definisi fungsi gaya sesuai soal 1
def F(x):
    return 5 + 2 * x**2

# Fungsi integrator metode titik tengah
def midpoint_rule(a, b, n):
    dx = (b - a) / n
    # Membuat array seluruh titik tengah: x_i* = a + (i - 0.5) * dx
    x_mid = a + (np.arange(1, n + 1) - 0.5) * dx
    return dx * np.sum(F(x_mid))

# Parameter awal dari soal
a = 0
b = 4
n = 4
tolerance = 1e-14

print("=== ITERASI KONVERGENSI METODE TITIK TENGAH ===")
print(f"{'n':<10}{'Hasil W (Joule)':<25}{'Selisih (Presisi)':<25}")

# Evaluasi awal untuk n = 4
W_old = midpoint_rule(a, b, n)
print(f"{n:<10}{W_old:<25.16f}{'-':<25}")

# Loop pembuktian konvergensi hingga presisi mesin
while True:
    n *= 2  # Menggandakan nilai n
    W_new = midpoint_rule(a, b, n)
    diff = abs(W_new - W_old)

    print(f"{n:<10}{W_new:<25.16f}{diff:<25.1e}")

    if diff < tolerance:
        print(f"\n[Selesai] Konvergen optimal tercapai pada n = {n}")
        print(f"Nilai Usaha (W) terbaik = {W_new:.16f} Joule")
        break

    W_old = W_new
```

```python
# Definisi fungsi kapasitas kalor sesuai soal 2
def C(T):
    return 20 + 0.05 * T**2

# Fungsi integrator metode trapesium
def trapezoidal_rule(a, b, n):
    dT = (b - a) / n
    # Penjumlahan komponen ujung: C(a) + C(b)
    Q_sum = C(a) + C(b)

    # Penjumlahan komponen dalam (interior) dikalikan 2
    for i in range(1, n):
        T_i = a + i * dT
        Q_sum += 2 * C(T_i)

    return (dT / 2) * Q_sum

# Parameter awal dari soal
a = 0
b = 100
n = 4
tolerance = 1e-14

print("=== ITERASI KONVERGENSI METODE TRAPESIUM ===")
print(f"{'n':<10}{'Hasil Q (Joule)':<25}{'Selisih (Presisi)':<25}")

# Evaluasi awal untuk n = 4
Q_old = trapezoidal_rule(a, b, n)
print(f"{n:<10}{Q_old:<25.16f}{'-':<25}")

# Loop pembuktian konvergensi
while True:
    n *= 2  # Menggandakan nilai n
    Q_new = trapezoidal_rule(a, b, n)
    diff = abs(Q_new - Q_old)

    print(f"{n:<10}{Q_new:<25.16f}{diff:<25.1e}")

    if diff < tolerance:
        print(f"\n[Selesai] Konvergen optimal tercapai pada n = {n}")
        print(f"Nilai Kalor (Q) terbaik = {Q_new:.16f} Joule")
        break

    Q_old = Q_new
```

```python
# Definisi fungsi arus listrik sesuai soal 3
def I_func(t):
    return 2 + 3 * t + t**2

# Fungsi integrator metode Simpson 1/3
def simpson_one_third(a, b, n):
    if n % 2 != 0:
        raise ValueError("Nilai n harus genap untuk metode Simpson 1/3!")

    dt = (b - a) / n
    q_sum = I_func(a) + I_func(b)

    # Loop untuk menerapkan koefisien 4 (ganjil) dan 2 (genap)
    for i in range(1, n):
        t_i = a + i * dt
        if i % 2 != 0:
            q_sum += 4 * I_func(t_i)
        else:
            q_sum += 2 * I_func(t_i)

    return (dt / 3) * q_sum

# Parameter awal dari soal
a = 0
b = 6
n = 6  # Harus genap
tolerance = 1e-14

print("=== ITERASI KONVERGENSI METODE SIMPSON 1/3 ===")
print(f"{'n':<10}{'Hasil q (Coulomb)':<25}{'Selisih (Presisi)':<25}")

# Evaluasi awal untuk n = 6
q_old = simpson_one_third(a, b, n)
print(f"{n:<10}{q_old:<25.16f}{'-':<25}")

# Loop pembuktian konvergensi
while True:
    n *= 2  # Menggandakan nilai n agar tetap genap
    q_new = simpson_one_third(a, b, n)
    diff = abs(q_new - q_old)

    print(f"{n:<10}{q_new:<25.16f}{diff:<25.1e}")

    if diff < tolerance:
        print(f"\n[Selesai] Konvergen optimal tercapai pada n = {n}")
        print(f"Nilai Muatan Listrik (q) terbaik = {q_new:.16f} Coulomb")
        break

    q_old = q_new
```
