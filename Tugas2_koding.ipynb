```python

import math
import pandas as pd


def bisection(f, a, b, tol=1e-10, max_iter=100):
    history = []
    for i in range(1, max_iter + 1):
        c = (a + b) / 2
        fa, fb, fc = f(a), f(b), f(c)

        history.append([i, a, b, c, fc])

        if abs(fc) < tol:
            return c, history

        if fa * fc < 0:
            b = c
        else:
            a = c

    return c, history


def regulafalsi(f, a, b, tol=1e-10, max_iter=100):
    history = []
    c_old = None

    for i in range(1, max_iter + 1):
        fa, fb = f(a), f(b)
        c = b - fb * (b - a) / (fb - fa)
        fc = f(c)

        ea = abs(c - c_old) if c_old else None

        history.append([i, a, b, c, fc, ea])

        if abs(fc) < tol:
            return c, history

        if fa * fc < 0:
            b = c
        else:
            a = c

        c_old = c

    return c, history


def secant(f, x0, x1, tol=1e-10, max_iter=100):
    history = []

    for i in range(1, max_iter + 1):
        f0, f1 = f(x0), f(x1)
        x2 = x1 - f1 * (x1 - x0) / (f1 - f0)

        ea = abs(x2 - x1)

        history.append([i, x0, x1, x2, ea])

        if ea < tol:
            return x2, history

        x0, x1 = x1, x2

    return x2, history


def newtonrap(f, df, x0, tol=1e-10, max_iter=100):
    history = []

    for i in range(1, max_iter + 1):
        fx = f(x0)
        dfx = df(x0)

        x1 = x0 - fx / dfx
        ea = abs(x1 - x0)

        history.append([i, x0, x1, ea])

        if ea < tol:
            return x1, history

        x0 = x1

    return x1, history
```

```python
def f(x):
  return x**3- x - 2

x1 = 1
xu = 2
max_iter = 10

print("Iter | x1 | xu | xr | f(xr) | Eror")
print("-"*65)

xr_old = 0

for i in range(1, max_iter+1):
  xr = xu - (f(xu)*(x1-xu))/(f(x1)-f(xu))

  if i == 1:
    eror = "-"
else :
  eror = abs((xr-xr_old)/xr)*100

  print(f"{i:4} | {x1:.6f} | {xu:.6f} | {xr:.6f} | {f(xr):.6f} | {eror}")

  if f(x1)*f(xr) < 0:
    xu = xr
  else:
    x1 = xr

  xr_old = xr
```

Metode Secant

```python
def f(x):
  return x**3 - x - 2

x0 = 1
x1 = 2
max_iter = 10
tol = 1e-10

print("Iter |      x0      |      x1      |      x2      |    f(x2)   |  Eror (ea)")
print("-"*85)

for i in range(1, max_iter + 1):
  f0 = f(x0)
  f1 = f(x1)

  if (f1 - f0) == 0:
      print(f"Warning: Division by zero encountered at iteration {i}. Stopping.")
      break

  x2 = x1 - f1 * (x1 - x0) / (f1 - f0)
  fx2 = f(x2)
  ea = abs(x2 - x1)

  print(f"{i:4} | {x0:12.6f} | {x1:12.6f} | {x2:12.6f} | {fx2:10.6f} | {ea:10.8f}")

  if ea < tol:
    break

  x0 = x1
  x1 = x2
```
